```python
#pie chart

import matplotlib.pyplot as plt
langs=["c","python","java"]
students=[20,100,40]
plt.pie(students,labels=langs)
plt.title("pie chart")
plt.show()
```

```python

import matplotlib.pyplot as plt
langs=["c","python","java"]
students=[20,100,40]
plt.pie(students,labels=langs,autopct="%1.1f%%") #to represent values in percentage
plt.title("pie chart")
plt.show()
```

```python
my_explode=(0.0,1.0,0) #give only one 0 more than one zero it gives error
import matplotlib.pyplot as plt
langs=["c","python","java"]
students=[20,100,40]
plt.pie(students,labels=langs,autopct="%1.1f%%",explode=my_explode, shadow=True) 
plt.title("pie chart")
plt.show()
```

my_explode=(0.0,1.0,0) was too many space between them so in next code we reduced to (0.0,0.1,0) in order to remove too many space between them

```python
my_explode=(0.0,0.1,0) #give only one 0 more than one zero it gives error
import matplotlib.pyplot as plt
langs=["c","python","java"]
students=[20,100,40]
plt.pie(students,labels=langs,autopct="%1.1f%%",explode=my_explode, shadow=True) 
plt.title("pie chart")
plt.show()
```
