```python
import pandas as pd

df["decision_date"] = pd.to_datetime(df["decision_date"])

df["year"] = df["decision_date"].dt.year
df["year"]
```

```python


fig,ax=plt.subplots(figsize=(10,8))
sns.countplot(x="year",hue="case_status",data=df)
ax.set(xlabel="visa application year",ylabel="No of applications")
plt.show()
```

```python
#insights as application increase, denied and accpeted both increasing 

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(10, 8))
sns.countplot(x="employer_city", hue="case_status", data=df, order=df["employer_city"].value_counts().iloc[:10].index)
ax.set(xlabel="Employer City", ylabel="Number of Applications")
plt.xticks(rotation=45)  # Rotating x-axis labels for better readability
plt.show()
```
