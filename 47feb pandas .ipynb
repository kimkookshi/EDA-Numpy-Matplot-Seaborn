```python
#merge and joint operation 
df1=pd.DataFrame({"key1":[1,2,4,5,6],
              "key2":[4,5,6,7,8],
              "key3":[3,4,5,6,6]})
df1
```

```python
df2=pd.DataFrame({"key1":[1,2,45,6,67],
                  "key4":[56,5,6,7,8],
                  "key5":[3,56,5,6,6]})
df2
```

```python
#key1 is common for  df1 and df2

#merge 
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how="left")
```

```python
#merge
pd.merge(df1,df2,how="right")
```

```python
#cross joint  #meaning : cartition product that is every element is mulyiplied or clubbing with every element
pd.merge(df1,df2,how="cross")
```

```python
pd.merge(df1,df2,how="left",left_on="key2",right_on="key4")
```

```python
#join
df1=pd.DataFrame({"key1":[1,2,4,5,6],
              "key2":[4,5,6,7,8],
              "key3":[3,4,5,6,6]})
index=["a","b","c","d","e"]
```

```python
df2=pd.DataFrame({"key1":[1,2,45,6,67],
                  "key4":[56,5,6,7,8],
                  "key5":[3,56,5,6,6]})
index=["a","b","h","i","j"]
```

```python
df1
```

```python
df2
```

```python
result = df1.join(df2, how="left", lsuffix='_df1', rsuffix='_df2')  #this lsuffix and rsuffix corrects the overlapping columns for df1 and df2
  #join happens on the indexes
result
```

```python
df1.join(df2,how="right",lsuffix='_df1', rsuffix='_df2')
```

```python
df1.join(df2,how="cross",lsuffix='_df1', rsuffix='_df2')
```
